```python
import tkinter as tk
import pandas as pd
import os

path = os.getcwd()+'/data'
print(path)
nameLs = os.listdir(path)

with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','r+') as f:
    num = int(f.read())
#print(num)




window = tk.Tk()  #创建窗口
window.title("好大夫标注器")  #窗口标题
window.geometry('1000x800')     #窗口大小，小写字母x
#窗口的label
textcontent = tk.StringVar() #文字变量储存
k = tk.Label(window,
    textvar = textcontent,  #文本,可变换
    bg ='white',   #字体的背景颜色
    font = ('Arial',12),  #字体和大小
    width = 30 , height = 2   #字体所占的宽度和高度
    )
k.pack()  #固定窗口
hit_state = False
i = 0
textcontent.set(i)     #为textcontent置值
def hit_here():
    global hit_state, i     #指定为全局变量，如果你想要为一个定义在函数外的变量赋值，
                            # 那么你就得告诉Python这个变量名不是局部的，而是 全局 的。
                            #我们使用global语句完成这一功能。没有global语句，是不可能为定义在函数外的变量赋值的。
    if hit_state == False:
        hit_state = True
        i+=1
        textcontent.set(i)
    else:
        hit_state = False
        i+=1
        textcontent.set(i)
#添加按钮
b = tk.Button(
    window,
    text = 'button',
    width = 5 , height = 2,
    command = hit_here   #执行函数体，而不是得到函数执行的结果
)
b.pack()
 
#以上是窗口的主体
window.mainloop() #结束（不停循环刷新）
```

```python
import tkinter
from  tkinter import ttk  #导入内部包
import pandas as pd
import os
import time

path = os.getcwd()+'/data'
print(path)
nameLs = os.listdir(path)

with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','r+') as f:
    num = int(f.read())
count = num


df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/6815737754--11.xls') 
df = df.fillna("无")

win=tkinter.Tk()
win.title("好大夫标注器")
win.geometry('1920x1080')


tree=ttk.Treeview(win)#表格
tree["columns"]=("医生姓名","医生文本","患者","患者文本")
tree.column("医生姓名",width=200)   #表示列,不显示
tree.column("医生文本",width=400)
tree.column("患者",width=50)
tree.column("患者文本",width=600)
 
tree.heading("医生姓名",text="医生姓名")  #显示表头
tree.heading("医生文本",text="医生文本")
tree.heading("患者",text="患者")
tree.heading("患者文本",text="患者文本")
 
for i in range(len(df.index)):   
    tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据
 
tree.pack()


def integrate():
    global tree,count
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','a+') as f:
        f.write(nameLs[count])
        f.write(',')
        f.write('integrate')
        f.write('\n')
        
    x = tree.get_children()
    for item in x:
        tree.delete(item)
        
    count+=1
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[count])
    df = df.fillna("无")
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','w+') as f:
        f.write(str(count))
    
    for i in range(len(df.index)):   
        tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据    
    tree.pack()
    

def helping():
    global tree,count
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','a+') as f:
        f.write(nameLs[count])
        f.write(',')
        f.write('help')
        f.write('\n')
        
    x = tree.get_children()
    for item in x:
        tree.delete(item)
        
    count+=1
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[count])
    df = df.fillna("无")
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','w+') as f:
        f.write(str(count))
    
    for i in range(len(df.index)):   
        tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据    
    tree.pack()

def individual():
    global tree,count
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','a+') as f:
        f.write(nameLs[count])
        f.write(',')
        f.write('individual')
        f.write('\n')
        
    x = tree.get_children()
    for item in x:
        tree.delete(item)
        
    count+=1
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[count])
    df = df.fillna("无")
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','w+') as f:
        f.write(str(count))
    
    for i in range(len(df.index)):   
        tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据    
    tree.pack()

def regret():
    global tree,count
    
    count-=1
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','r+') as f:
        lines = f.readlines()
        lines = lines[:-1]
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','w+') as f:
        for line in lines:
            f.write(line)
            f.write('\n')
    
    x = tree.get_children()
    for item in x:
        tree.delete(item)
        
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[count])
    df = df.fillna("无")
    
    for i in range(len(df.index)):   
        tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据   
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','w+') as f:
        f.write(str(count))
    tree.pack()

def noAnswer():
    global tree,count
    
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/resultsLog.txt','a+') as f:
        f.write(nameLs[count])
        f.write(',')
        f.write('noAnswer')
        f.write('\n')
        
    x = tree.get_children()
    for item in x:
        tree.delete(item)
        
    count+=1
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[count])
    df = df.fillna("无")
    with open('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/tkinterLog.txt','w+') as f:
        f.write(str(count))
    
    for i in range(len(df.index)):   
        tree.insert("",i,text=str(i),values=(df.iloc[i,0],df.iloc[i,1],df.iloc[i,2],df.iloc[i,3])) #插入数据    
    tree.pack()
        
#添加按钮
aButton = tkinter.Button(
    win,
    text = '集成协作模式',
    width = 15 , height = 2,
    command = integrate   #执行函数体，而不是得到函数执行的结果
)
aButton.pack()

bButton = tkinter.Button(
    win,
    text = '助理协作模式',
    width = 15 , height = 2,
    command = helping   #执行函数体，而不是得到函数执行的结果
)
bButton.pack()

cButton = tkinter.Button(
    win,
    text = '独立协作模式',
    width = 15 , height = 2,
    command = individual   #执行函数体，而不是得到函数执行的结果
)
cButton.pack()

backButton = tkinter.Button(
    win,
    text = '选错了，我要反悔',
    width = 15 , height = 2,
    command = regret   #执行函数体，而不是得到函数执行的结果
)
backButton.pack()

noAnswerButton = tkinter.Button(
    win,
    text = '空对话，或无回答',
    width = 15 , height = 2,
    command = noAnswer   #执行函数体，而不是得到函数执行的结果
)
noAnswerButton.pack()



win.mainloop()
```
