```python
def click_by_time(driver,xpath,maxTime):#防止动态元素未加载出来，无法点击导致程序出错，故定义该方法，等元素加载出来再点击它
    t = 0
    while t<=maxTime:
        if driver.find_element_by_xpath(xpath)!=None:
            driver.find_element_by_xpath(xpath).click()
            break
        time.sleep(1)
        t+=1
```

```python
from selenium import webdriver
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import xlwt

with open('/Users/fowillwly/Documents/SHU/2020夏季学期/好大夫在线爬虫/convUrl.txt','r+') as f:
    ls = []
    for url in f.readlines():
        ls.append('https://m.haodf.com/'+url[:-1])

team = 'https://m.haodf.com/doctorteam/7367388699.htm'
count = 0

for i in range(len(ls)):
    
    url = ls[i]
    driver = webdriver.Chrome()
    driver.get(url)

    while driver.find_element_by_xpath('/html/body/section[3]/a').text[-4:] == '加载更多':
        button = driver.find_element_by_xpath('/html/body/section[3]/a')
        driver.execute_script("arguments[0].click();", button)
        time.sleep(2)
    
    teamUrl = driver.find_element_by_xpath('/html/body/div[8]/div/div/div/a[1]').get_attribute('href')[-14:-4]
    print(teamUrl)
    
    if teamUrl == team:
        count += 1
    else:
        team = teamUrl
        count = 1
    
    workbook = xlwt.Workbook(encoding = 'utf-8')
    worksheet = workbook.add_sheet('好大夫')
    worksheet.write(0,0,'医生姓名')
    worksheet.write(0,1,'医生文本')
    worksheet.write(0,2,'患者')
    worksheet.write(0,3,'患者文本')
    workbook.save('/Users/fowillwly/Documents/SHU/2020夏季学期/好大夫在线爬虫/data/'+teamUrl+'--'+str(count)+'.xls')
    
    doctors = driver.find_elements_by_xpath('//*[@class="ele-left"]/div[1]/div/p')
    for m in range(len(doctors)):
        worksheet.write(m+1,0,doctors[m].text)
        
    doctorsTexts = driver.find_elements_by_xpath('//*[@class="ele-left"]/div[2]/p[1]')
    for m in range(len(doctorsTexts)):
        worksheet.write(m+1,1,doctorsTexts[m].text)

    patents = driver.find_elements_by_xpath('//*[@class="ele-right"]/div[1]/div/p')
    for m in range(len(patents)):
        worksheet.write(m+1,2,patents[m].text)
        
    patentsTexts = driver.find_elements_by_xpath('//*[@class="ele-right"]/div[2]/p[1]')
    for m in range(len(patentsTexts)):
        worksheet.write(m+1,3,patentsTexts[m].text)
        
    workbook.save('/Users/fowillwly/Documents/SHU/2020夏季学期/好大夫在线爬虫/data/'+teamUrl+'--'+str(count)+'.xls')  
    
    driver.quit()
    print(i)
                                             
```
