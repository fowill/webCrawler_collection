```python
path = os.getcwd()+'/data'
print(path)
```

```python
nameLs = os.listdir(path)
```

```python
#需要建立一个dict，每个key是专家团队的名称，value是一个list，包括：该专家团队摘取的对话数，专家人数，专家姓名(list)，专家回答句数，
#专家平均回话字数，用户回答句数，用户平均问话字数，

#然后，转化成xls文件
```

```python
import pandas as pd
import xlwt
```

```python
numDict = {}
for name in nameLs:
    if numDict.get(name,'no') == 'no':
        numDict[name] = 1
    else:
        numDict[name] += 1
print(numDict)
```

```python
workbook = xlwt.Workbook(encoding = 'utf-8')
worksheet = workbook.add_sheet('好大夫在线')
worksheet.write(0,0,'团队id')
worksheet.write(0,1,'参与本次对话医生数')
worksheet.write(0,2,'医生姓名')
worksheet.write(0,3,'医生回复句数')
worksheet.write(0,4,'医生平均回复字数')
worksheet.write(0,5,'用户询问句数')
worksheet.write(0,6,'用户平均询问字数')
workbook.save('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/singleConvData.xls')

for i in range(len(nameLs)):
    df = pd.read_excel('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/data/'+nameLs[i])
    df = df.fillna('None')
    df = df.replace('好大夫在线友情提示：请详细描述或拍照上传病历资料，以便医生了解病情，做出更好的诊断。同时，线上咨询不同于线下面诊，医生的建议仅供参考。','')

    doctorDF = df[df['医生姓名']!='None']
    numOfDoctor = len(set(doctorDF['医生姓名']))
    doctors = list(set(doctorDF['医生姓名']))
    nameOfDoctor = ''
    for doctor in doctors:
        nameOfDoctor = nameOfDoctor + '/' +str(doctor)
    numOfDoctorSentance = len(list(doctorDF['医生姓名']))
    numOfDoctorWord = doctorDF['医生文本'].str.len().mean()

    patentDF = df[df['患者']!='None']
    numOfPatentSentance = len(list(patentDF['患者']))
    numOfPatentWord = patentDF['患者文本'].str.len().mean()
    
    worksheet.write(i+1,0,nameLs[i][:10])
    worksheet.write(i+1,1,numOfDoctor)
    worksheet.write(i+1,2,nameOfDoctor)
    worksheet.write(i+1,3,numOfDoctorSentance)
    worksheet.write(i+1,4,numOfDoctorWord)
    
    worksheet.write(i+1,5,numOfPatentSentance)
    worksheet.write(i+1,6,numOfPatentWord)
    
    workbook.save('/Users/fowillwly/Dev/webCrawler_collection/GoodDoctorWebCrawler/singleConvData.xls')
    
    print(i,len(nameLs))
```
